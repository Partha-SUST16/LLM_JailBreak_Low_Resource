```python
import pandas as pd

# List of language codes
language_codes = ['hi', 'ur', 'bn', 'ne', 'si', 'ta', 'dz', 'dv', 'ps', 'en']

# Define the model
model = "llama/llama3-70b-8192"
model = "mixtral/mixtral-8x7b-32768"
model = "deepseek/deepseek"
model = "gemma/gemma"
# Initialize an empty list to store dataframes
df_list = []

# Loop through each language code and load the CSV
for language_code in language_codes:
    file_path = f'/content/drive/MyDrive/CPSC 538P Project (Computer Security)/Evaluation/Qwen/{model}_{language_code}_labels.csv'

    # Read the CSV file
    df = pd.read_csv(file_path)

    # Add a new column for language code
    df['language_code'] = language_code

    # Clean classification column (remove quotes if present)
    df['classification'] = df['classification'].str.replace('"', '')
    df['classification'] = df['classification'].str.replace('[', '')
    df['classification'] = df['classification'].str.replace(']', '')
    df['classification'] = df['classification'].str.replace('\'', '')
    # Assign label 1 if classification is "following", else 0
    df['label'] = (df['classification'].notna() & df['classification'].str.contains("follow", na=False)).astype(int)#(df['classification'] == "following").astype(int)
    df['model'] = model.split('/')[0]
    # Append to the list
    df_list.append(df)

# Concatenate all dataframes
merged_df = pd.concat(df_list, ignore_index=True)

# Save the merged dataframe (optional)
merged_df.to_csv('/content/drive/MyDrive/CPSC 538P Project (Computer Security)/Evaluation/Qwen/merged_labels_'+model.split('/')[0]+'.csv', index=False)

# Display the first few rows
print(merged_df.head())
```

```python
import pandas as pd

# List of models
models = ["llama/llama3-70b-8192", "mixtral/mixtral-8x7b-32768", "deepseek/deepseek", "gemma/gemma"]

# Initialize an empty list to store dataframes
df_list = []

# Loop through each model and load its merged CSV
for model in models:
    model_name = model.split('/')[0]  # Extract base model name
    file_path = f'/content/drive/MyDrive/CPSC 538P Project (Computer Security)/Evaluation/Qwen/merged_labels_{model_name}.csv'

    # Read the CSV file
    df = pd.read_csv(file_path)

    # Add a new column for the model name
    df['model'] = model_name

    # Append to the list
    df_list.append(df)

# Concatenate all dataframes
final_merged_df = pd.concat(df_list, ignore_index=True)

# Save the final merged dataframe
final_merged_df.to_csv('/content/drive/MyDrive/CPSC 538P Project (Computer Security)/Evaluation/Qwen/final_merged_labels.csv', index=False)

# Display the first few rows
print(final_merged_df.head())
```
